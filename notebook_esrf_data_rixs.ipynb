Extract and save RIXS spectra from a .spec file

```python
"""
This code processes RIXS spectra from a specified folder and a list of runs using the esrf_run_spectrum class.

Parameters:
-   folder (str): The path to the folder containing the data files.
-   runs (list): A list of run numbers to process.
-   points (list): A list of points to extract from each run. 
    Each element of the list is a list of points to extract from the corresponding run. 
    If 'all' is specified, all points are extracted.
-   x_name (str): The name of the x-axis variable.
-   y_name (str): The name of the y-axis variable.
-   norm_name (str): The name of the normalization variable.
-   motor_names (list): A list of motor names to include in the processing.

The esrf_run_spectrum object is created with the specified folder and runs. The process_spectra method is then called on this object with the specified parameters to process the spectra and return the results as an xarray Dataset.
"""

folder = r"C:\Users\leona\OneDrive - Universität Zürich UZH\HC5868\data\trial"
runs = [65,66]
scans = [['all'],['all']]
x_name = 'Pixel'
y_name = 'SPC'
norm_name = 'mir_xmcd'
motor_names = ['srot', 'magnet', 'energy']
scans_from_same_run = True
sample_name = "TmIG"
file_path_save = r"C:\Users\leona\OneDrive - Universität Zürich UZH\HC5868\data\extracted\leo\trial.hdf5"

run_spec = esrf_run_spectrum(folder, runs, scans)
xarrays = run_spec.extract_and_process_spectra(x_name,
                                    y_name,
                                    norm_name, 
                                    motor_names,
                                    sample_name=sample_name,
                                    fit_shifts=False,
                                    align_spectra=True,
                                    correlation_batch_size=400,
                                    poly_order=1,
                                    plot=True,
                                    scans_from_same_run=scans_from_same_run)

run_spec.save_to_hdf5(file_path_save)
run_spec.save_to_csv_for_originlab(file_path_save.replace('.hdf5', '.csv'),
                                   motor_names=['srot', 'energy'],
                                   motor_name_mapping={'theta': 'srot', 'energy': 'energy'},)



```
